# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## Do you agree with the above analysis and recommendations? Why or why not?

This initial analysis provides a good first take using Exploratory Data Analysis (EDA) but does not dive into the statistical inference to verify the conclusions from the EDA. I do not agree with the above analysis because we don't know for sure if the number of discharges is actually correlated with the excess readmission rate. What is the correlation value? Is this statistically significant? 
<br>
<br>
We also need to take a look at whether the excess readmission rate mean of hospitals with lower discharges is different from hopitals with a more discharges.
<br>
<br>
The first thing we will need to look as is determine what the correlation is between discharges and the excess readmission rate.

```python
clean_hospital_read_df.isnull().sum()
```

```python
hosp_cleaner = clean_hospital_read_df[clean_hospital_read_df['Excess Readmission Ratio'].notnull()]
hosp_cleaner = hosp_cleaner.reset_index()
hosp_cleaner.isnull().sum()
```

```python
corr_dis_read = np.corrcoef(hosp_cleaner['Number of Discharges'], hosp_cleaner['Excess Readmission Ratio'])
corr = corr_dis_read[0,1]
print("Correlation between number of discharges and excess readmission ratio: " + str(corr))
```

```python
r_squared = corr ** 2
print("The r-squared value for number of discharges and excess readmission ratios: " + str(r_squared))
```

The correlation for the number of discharges and the excess readmission ratio is -0.0974 which means that they are negatively associated (discharge goes up then excess readmission ratio goes down). However, the correlation is not very high. Correlation can be from -1 to 1 with highly correlated values being closer to -1 or 1. The next step is to perform a hypothesis test to see if the correlation is statistically significant.
<br>
<br>
## Practical Significance vs Statistical Significance
Statistical significance in this case shows that the correlation between number of discharges and excess readmission ratio is more than just random chance. This is the case because of how many samples were used in the test (a t-test usually requires more than 20 samples).
<br>
However, practical significance looks at how big this correlation is and whether it actually means something in the real world. In this case, the correlation is only -.09, which is very small since a value of zero (0) means two variables are not correlated. 
<br>
For the correlation, practical significance is more important in this case. There is a correlation between the two variables but the value is small.
<br>
<br>
## Additional Hypothesis Tests
In addition to the hypothesis test for the correlation variable, more tests need to be done on the ratio values for the hospitals with different discharges and number of discharges based on different ratios.
<br>
<br>
**Null Hypothesis**: There is no correlation between the number of discharges and the excess readmission ratio.
<br>
**Alternative Hypothesis**: There is a correlation between the number of discharges and the excess readmission ratio.
<br> 
**Significance level**: 0.01
<br>
<br>
The test used here will be a Student's T-Test because the actual standard deviation of the correlation is not known.

```python
t_num = corr * (np.sqrt(len(hosp_cleaner) - 2))
t_denom = np.sqrt((1 - (corr ** 2)))
t_stat = t_num / t_denom
pval = stat.t.sf(np.abs(t_stat), len(hosp_cleaner) - 1)*2
print("The p-value for the t-test for the correlation between discharges and excess readmission ratios: " + str(pval))
```

This p-value is less than the significance level of 0.01 so we can reject the null hypothesis that there is no correlation between the two variables. However, the correlation is not very high and the r-squared value is 0.949%. This means that only 0.949% of the data match the actual regression line of the data.
<br>
## Hypothesis Test for Excess Readmission Ratios
For this hypothesis test, I will split the data up between small (< 100), medium (between 100 and 1000), and large (excess of 1000) number of discharges and compare the means. If there is a difference in the means, I will then perform additional t-tests to determine which means are different.
<br>
<br>
**Null Hypothesis**: The means for all three groups are equal.
<br>
**Alternative Hypothesis**: The means for the three groups are not equal.
<br>
**Statistical Significance**: 0.01

```python
small = hosp_cleaner[hosp_cleaner['Number of Discharges'] < 100]['Excess Readmission Ratio']
medium = hosp_cleaner[(hosp_cleaner['Number of Discharges'] >= 100) & (hosp_cleaner['Number of Discharges'] <= 1000)]['Excess Readmission Ratio']
large = hosp_cleaner[hosp_cleaner['Number of Discharges'] > 1000]['Excess Readmission Ratio']

large_mean = np.mean(large)
medium_mean = np.mean(medium)
small_mean = np.mean(small)

print("Mean Excess Readmission Ratios for hospitals with discharges less than 100: " + str(large_mean))
print("Mean Excess Readmission Ratios for hospitals with discharges between 100 and 1000: " + str(medium_mean))
print("Mean Excess Readmission Ratios for hospitals with discharges greater than 1000: " + str(small_mean))
```

The mean excess readmission ratios for the different number of discharges are all different, though they are small differences. An analysis of variance test needs to be performed to determine if the differences are statistically significant.

```python
anova = stat.f_oneway(small, medium, large)
p_val = anova[1]
print("The p-value for the ANOVA test: " + str(p_val))
```

The p_value for the ANOVA test is less than 0.01 so we can reject the null hypothesis that there is no difference between the means of the two groups. Because of this, we know that there is at least one difference in the means. To determine which means are different, t-tests will be performed.
<br>
<br>
Because multiple t-tests will be completed, we'll need to correct the significance of the test because significance is lost. We'll use the Bonferroni correction which is alpha divided by the number of tests (3).

```python
new_alpha = 0.01 / 3
print("Bonferroni correction for the significance level: " + str(new_alpha))
```

**Null Hypothesis**: The means for the small number of discharges and the medium discharges are not different
<br>
**Alternative Hypothesis**: The means for the small number of discharges and the medium discharges are different

```python
t_small_med = stat.ttest_ind(small, medium, equal_var = False)
p_val = t_small_med[1]
print("The p-value for the t-test between the differences in means of ratios of small and medium discharges: " + str(p_val))
```

The p-value in this t-test is lower than the significance level, therefore we can reject the null hypothesis that there is no difference in ratios between hospitals with less than 100 discharges and hospitals with discharges between 100 and 1000.

**Null Hypothesis**: The means for the small number of discharges and the large discharges are not different
<br>
**Alternative Hypothesis**: The means for the small number of discharges and the large discharges are different

```python
t_small_large = stat.ttest_ind(small, large, equal_var = False)
p_val_l = t_small_large[1]
print("The p-value for the t-test between the differences in means of ratios of small and large discharges: " + str(p_val_l))
```

The p-value in this t-test is lower than the significance level, therefore we can reject the null hypothesis that there is no difference in ratios between hospitals with less than 100 discharges and hospitals with more than 1000 discharges.

**Null Hypothesis**: The means for the small number of discharges and the medium discharges are not different
<br>
**Alternative Hypothesis**: The means for the small number of discharges and the medium discharges are different

```python
t_med_large = stat.ttest_ind(medium, large, equal_var = False)
p_val_m = t_med_large[1]
print("The p-value for the t-test between the differences in means of ratios of small and large discharges: " + str(p_val_m))
```

The p-value in this t-test is lower than the significance level, therefore we can reject the null hypothesis that there is no difference in ratios between hospitals with more than 1000 discharges and hospitals with discharges between 100 and 1000.
<br>
<br>
In addition to testing the means, we can also test whether the number of discharges are different for ratios less than or equal to 1 and ratios greater than 1.

```python
rate_under_1 = hosp_cleaner[hosp_cleaner['Excess Readmission Ratio'] <= 1][['Number of Discharges', 'Excess Readmission Ratio']]
rate_over_1 = hosp_cleaner[hosp_cleaner['Excess Readmission Ratio'] > 1][['Number of Discharges','Excess Readmission Ratio']]

rate_u1_mean = rate_under_1['Number of Discharges'].mean()
rate_o1_mean = rate_over_1['Number of Discharges'].mean()

print("Mean Number of Discharges for Hospitals with an excess readmission ratio equal to 1: " + str(rate_u1_mean))
print("Mean Number of Discharges for Hospitals with an excess readmission ratio greater than or equal to 1: " + str(rate_o1_mean))
print("Difference in means: " + str(rate_u1_mean - rate_o1_mean))
```

**Null Hypothesis**: The means of the number of discharges for hospitals with an excess readmission rate less than 1 and ratios greater than or equal to 1 are not different.
<br>
**Alternative Hypothesis**: The means of the number of discharges for hospitals with an excess readmission rate less than 1 and ratios greater than or equal to 1 are different.
**Significance Level**: 0.01

```python
t_rates = stat.ttest_ind(rate_under_1, rate_over_1, equal_var = False)
pval = t_rates[1]
print("P-value for the difference in number of discharges means: " + str(pval))
```

The p-value here is less than the significance level therefore we can reject the null hypothesis and conclude that there are differences in means between the number of discharges.
